```python
from pyvox.parser import VoxParser
import os
import numpy as np
import plotly.graph_objects as go
# import matplotlib.pyplot as plt

# os.getcwd()
# 'e:\\pku_semesters\\year2_fall\\CV_PengShuaiWang\\8_finalProject\\MakePottery\\utils'
s = "AL_13B-n012-t1649436923"
voxel_matrix = VoxParser(f"../data/train/1/{s}.vox").parse().to_dense()

import numpy as np
import plotly.graph_objects as go

def vox_to_numpy(vox_data):
    voxel_matrix = np.zeros((32, 32, 32), dtype=bool)
    for voxel in vox_data:
        x, y, z = voxel['coords']
        voxel_matrix[x, y, z] = True
    return voxel_matrix

import numpy as np
import plotly.graph_objects as go

def visualize_voxel_matrix(voxel_matrix):
    # 创建一个字典来存储每个值的颜色
    colors = ['white','#ceabb2', '#d05d86', '#7e1b2f', '#c1375b', '#cdc1c3',
              '#e377c2', '#7f7f7f', '#bcbd22', '#1f77b4', '#9edae5',
              '#d62728', '#ff7f0e', '#8c564b', '#56b4e9', '#c7c7c7']

    x, y, z = np.where(voxel_matrix)
    
    # 创建一个列表来存储每个散点的颜色
    marker_colors = []

    # 遍历数组中的每个元素，将每个元素的颜色添加到列表中
    for i in range(voxel_matrix.shape[0]):
        for j in range(voxel_matrix.shape[1]):
            for k in range(voxel_matrix.shape[2]):
                if voxel_matrix[i, j, k]:
                    marker_colors.append(colors[voxel_matrix[i, j, k]])

    # 使用scatter3d函数可视化数组
    fig = go.Figure(data=go.Scatter3d(x=x, y=y, z=z, mode='markers', marker=dict(size=5, symbol='square', color=marker_colors, line=dict(width=2, color='DarkSlateGrey',))))
    fig.update_layout(scene=dict(xaxis_title='X', yaxis_title='Y', zaxis_title='Z'), legend=dict(orientation="h"))
    fig.show()


# 示例：将VOX数据转换为np.ndarray类型，并使用plotly可视化
visualize_voxel_matrix(voxel_matrix)
```

```python
import numpy as np

# 使用histogram函数构建直方图
hist, bins = np.histogram(voxel_matrix, bins=np.arange(0, np.max(voxel_matrix)) + 0.5)

# 输出直方图
print("Bins:", bins)
print("Histogram:", hist)
```
